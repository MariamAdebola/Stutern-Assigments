```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
#read csv file
df = pd.read_csv('company_sales_data.csv')

#check
df.head()
```

Question1: Compare any two products sales data using a bar chart.

```python
ax = plt.subplot()

month_name = ['Jan', 'Feb', 'March', 'April', 'May', 'June', 'July', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
month = df['month_number']
facecream = df['facecream']
facewash = df['facewash']

xname = np.arange(len(month_name))

plt.bar(xname - 0.2, facecream, 0.5, color='g', label="facecream")
plt.bar(xname + 0.2, facewash, 0.5, color='b', label="facewash")

plt.xlabel("month")
plt.ylabel('product sales')
plt.title("Sales for Face Products")

plt.xticks(xname, month_name)
ax.set_xticklabels(month_name, rotation=90, fontsize=11)
plt.legend()
plt.show()
plt.savefig('faceproductsdata.png')
```

Question 2: Assume the average monthly sales of bathing soap is 10k, visualize the sales data of bathing soap of all months, showing how they differ from the average sale.

```python
month_name = ['Jan', 'Feb', 'March', 'April', 'May', 'June', 'July', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
bathingsoap = df['bathingsoap']

xname = np.arange(len(month_name))
yerr = list(list(map(lambda x: 10000-x, bathingsoap)))

plt.bar(xname, bathingsoap, color='b', label="facecream", yerr=yerr, capsize=10)
plt.xlabel("month")
plt.ylabel('soap sales')
plt.title("Monthly Sales for Bathing Soap")

plt.xticks(xname, month_name)
plt.show()
```

Question3: Calculate the total sales data for the year of each product and represent it on a pie chart

```python
fig = plt.figure(figsize =(5, 5))

products = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']
# df.sort_values(by=['Brand'], inplace=True, ascending=False)

SalesData = (df['facecream'].sum(), df['facewash'].sum(), df['toothpaste'].sum(), df['bathingsoap'].sum(),
df['shampoo'].sum(), df['moisturizer'].sum())

plt.pie(SalesData, labels=products, autopct='%d%%', wedgeprops=dict(width=0.55), colors=['k', 'g', 'b', 'y', 'm', 'c'])
plt.title("Total Sales per Product")
plt.axis('equal')
plt.savefig('Total Sales per Product.png')
plt.show()
```
