```python
#import libraries
from matplotlib import pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
#read csv file
df = pd.read_csv('anscombe_quartet_data.csv')

#check
df.head()
```

Method 1: plotting subplots

```python
fig = plt.figure(figsize=(10, 10))

x = range(0, 25)
y = range(0, 25)

x1 = df["x1"]
y1 = df["y1"]
plt.subplot(2, 2, 1)
plt.title('x1y1')
plt.xlabel('x1')
plt.ylabel('y1')
plt.scatter(x1, y1, color="green")
plt.plot(x, y)

x2 = df["x2"]
y2 = df["y2"]
plt.subplot(2, 2, 2)
plt.title('x2y2')
plt.xlabel('x2')
plt.ylabel('y2')
plt.scatter(x2, y2, color='grey')
plt.plot(x, y)

x3 = df["x3"]
y3 = df["y3"]
plt.subplot(2, 2, 3)
plt.title('x3y3')
plt.xlabel('x3')
plt.ylabel('y3')
plt.scatter(x3, y3, color='blue')
plt.plot(x, y)

x4 = df["x4"]
y4 = df["y4"]
plt.subplot(2, 2, 4)
plt.title('x4y4')
plt.xlabel('x4')
plt.ylabel('y4')
plt.scatter(x4, y4, color='black')
plt.plot(x, y)

plt.subplots_adjust(wspace=0.6, hspace=0.5, bottom=0.1, top=0.9)
plt.show()
```

Method 2: plotting subplots

```python
x = range(0, 25)
y = range(0, 25)

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(9, 9))

x1 = df["x1"]
y1 = df["y1"]
ax1.scatter(x1, y1, color="green")
ax1.plot(x, y)
ax1.set_title('x1y1')

x2 = df["x2"]
y2 = df["y2"]
ax2.scatter(x2, y2, color="green")
ax2.plot(x, y)
ax2.set_title('x2y2')

x3 = df["x3"]
y3 = df["y3"]
ax3.scatter(x3, y3, color="blue")
ax3.plot(x, y)
ax3.set_title('x3y3')

x4 = df["x4"]
y4 = df["y4"]
ax4.scatter(x4, y4, color="black")
ax4.plot(x, y)
ax4.set_title('x4y4')

plt.subplots_adjust(wspace=0.6, hspace=0.5, bottom=0.1, top=0.9)
```

Method 3: plotting subplots

```python
x = range(0, 25)
y = range(0, 25)

fig = plt.figure(figsize=(8, 6))

x1 = df["x1"]
y1 = df["y1"]
ax1= fig.add_subplot(221, title="x1y1")
ax1.scatter(x1, y1, color="green")
ax1.plot(x, y)

x2 = df["x2"]
y2 = df["y2"]
ax2= fig.add_subplot(222, title="x2y2")
ax2.scatter(x2, y2, color="green")
ax2.plot(x, y)

x3 = df["x3"]
y3 = df["y3"]
ax3 = fig.add_subplot(223, title="x3y3")
ax3.scatter(x3, y3, color="blue")
ax3.plot(x, y)

x4 = df["x4"]
y4 = df["y4"]
ax4 = fig.add_subplot(224, title="x4y4")
ax4.scatter(x4, y4, color="black")
ax4.plot(x, y)

plt.subplots_adjust(wspace=0.6, hspace=0.5, bottom=0.1, top=0.9)
```
