```python
from matplotlib import pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
#read csv file
df = pd.read_csv('company_sales_data.csv')

#check
df.head()
```

```python
x_axis = df["month_number"]

product1 = df["facecream"]
plt.plot(x_axis, product1, ls="--", marker="o", color="green")
plt.legend(['month_number', 'facecream'])

product2 = df["facewash"]
plt.plot(x_axis, product2, ls=":", marker="*", color="blue")

product3 = df["toothpaste"]
plt.plot(x_axis, product3, ls="--", marker="s", color="yellow")

product4 = df["bathingsoap"]
plt.plot(x_axis, product4, ls=":", marker="o", color="black")

product5 = df["shampoo"]
plt.plot(x_axis, product5, ls="--", marker="s", color="grey")

product6 = df["moisturizer"]
plt.plot(x_axis, product6, ls=":", marker="*", color="red")

plt.ylabel("total_units_sold")
plt.xlabel("month_number")
plt.title("Trends of Goods sold")
plt.legend(['facecream', 'toothpaste', 'facewash', 'bathingsoap', 'shampoo', 'moisturizer'], loc =0)
plt.show()
```
