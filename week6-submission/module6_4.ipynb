**Question 1: Use inner merge operation for the orders and products table**



```python
df_OrderProduct = pd.merge(df_order, df_product, how="inner")
df_OrderProduct
```

**My Observation:** *Without specifying the column to merge both dataframes, all data elements for each product_id present in both dataframes are returned.*

**Question 2: Merge OrderProduct with customer dataframe**

```python
df_OrderProductUpdated = pd.merge(df_OrderProduct, df_customer, how="inner")
df_OrderProductUpdated
```

**My Observation:** *Without specifying the column to merge both dataframes, all data elements for each customer_id present in both dataframes are returned.*

**Question 3: Use inner merge customers and order table** *(specify the column for the inner merge and use merge function arguments left_on, right_on and suffixes)*

```python
df_CustomerOrder = df_customer.merge(df_order, on="customer_id", how="left", suffixes=('_left', '_right'))
df_CustomerOrder
```

**My Observation:** After merging the two dataframes using a unique identifier (customer_id), all data values for each customer_id in the two dataframes are returned.

4. **Question 4: Use outer merge customers and product table** *(use the .merge argument how) and state your observations in a text cell.*

```python
df_CustomerProduct = df_customer.merge(df_product, left_index=True, right_index=True, how="outer", indicator=True)
df_CustomerProduct
```

**My Observation:** After merging the two dataframes with the outer merge, all data elements in the two dataframes are returned. Cells with NAN represent data points with no data values in the first dataframe.

**Question 5: Use left merge orders and customers table** *(use the .merge argument how)*

```python
df_CustomerOrderUpdated = pd.merge(df_order, df_customer, how="left", on="customer_id", indicator=True)
df_CustomerOrderUpdated
```

**My Observation:** After merging the two dataframes, all data elements in the first dataframe, including data elements in the second dataframe that has matching customer id with the first dataframe are returned. Cells with NAN represent data points with no data values in the second dataframe.

**Question 6:** Use right merge products and customers table (use the .merge argument how) and state your observations in a text cell.

```python
df_ProductCustomerUpdated = pd.merge(df_product, df_customer, left_index=True, right_index=True, how="right", indicator=True)
df_ProductCustomerUpdated
```

**My Observation:** After merging the two dataframes, all data elements in the second dataframe, including data elements in the first dataframe that has matching customer id with the first dataframe are returned. Cells with NAN represent data points with no data values in the first dataframe.
